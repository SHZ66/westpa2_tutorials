```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1 = np.loadtxt('trial_1/dataset/MFPT_convergence_no_err.dat')
t2 = np.loadtxt('trial_2/dataset/MFPT_convergence_no_err.dat')
t3 = np.loadtxt('trial_3/dataset/MFPT_convergence_no_err.dat')
```
